```python
def get_first3(row):
    if pd.isna(row['First_5']):
        return
    aid = row['Article ID']
    if pd.isna(row['[Cleansed - DU] PO text']):
        if pd.isna(row['[Cleansed - DU] Long desc']):
            if pd.isna(row['[Cleansed - DU] Short desc']):
                return
            else:
                desc = row['[Cleansed - DU] Short desc']
        else:
            desc=row['[Cleansed - DU] Long desc']
    else:
        desc=row['[Cleansed - DU] PO text']
    words = " ".join(desc.split(" ")[:3])
    grs = " ".join([gram[gr.upper()] for gr in row['Grammaire'].split(" ")[:3]])
    sys = " ".join([infos[sy.upper()] for sy in row['Syntax'].split(" ")[:3]])
    return [aid, words,grs,sys]
```

```python
def inv(morph):
    ls = morph.split("][")
    cpt=1
    while(cpt>0):
        cpt=0
        for idx,elem in enumerate(ls):
            elem=elem.replace("[","").replace("]","")
            ls[idx]=elem
            if len(elem)<3:
                if idx>0 and elem!='af':
                    ls[idx-1]=ls[idx-1]+ls[idx]
                    del ls[idx]
                    cpt+=1
                else:
                    try:
                        ls[idx+1]=ls[idx]+ls[idx+1]
                        del ls[idx]
                    except:
                        return ls
    sufs=['ing','aar','tor','tje','tie','ator','atie','atje']
    for suf in sufs:
        try:
            i=ls.index(""+suf+"]")
            ls[i-1]=ls[i-1]+""+suf+""
            del ls[i]
        except:
            pass

        try:
            i=ls.index(""+suf+"s]")
            ls[i-1]=ls[i-1]+""+suf+"s"
            del ls[i]
        except:
            pass

        try:
            i=ls.index(""+suf+"s")
            ls[i-1]=ls[i-1]+""+suf+"s"
            del ls[i]
        except:
            pass

        try:
            i=ls.index(""+suf+"")
            ls[i-1]=ls[i-1]+""+suf+""
            del ls[i]
        except:
            pass
    
    return list(ls)
```
