```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
from frog import FrogOptions,Frog
frog = Frog(FrogOptions(parser=True))
```

```python
import pandas as pd
df = pd.read_csv('MD6 - Cleansed descriptions - DU.csv')
```

```python
df.head()
```

```python
def process(desc):
    if(pd.isna(desc)):
        return
    desc=desc.lower()
    Lemma=[]
    for word in frog.process(desc):
       Lemma.append(word["lemma"]) 
    return " ".join([word.upper() for word in Lemma])
```

```python
for col in ['- DU] Short desc','- DU] Long desc','- DU] PO text']:
    df.to_csv(col[-6:])
    print('[Cleansed '+col)
    df['[Cleansed_SING_MASC '+col] = df['[Cleansed '+col].apply(process)
```

```python
df.to_csv('MD6 - Cleansed descriptions - DU_ WITH_LEMMA.csv')
```
