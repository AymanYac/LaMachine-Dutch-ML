```python
import unidecode
def keepalpha(word):
    #remove contiguous non alpha chars
    if not isinstance(word,str):
        return ""
    word= unidecode.unidecode(word)
    for idx,char in enumerate(word):
        offset=0
        ##if idx != 0 and not char.isdigit():
        ##    offset =1
        if not char.isalpha() :
            return word[:idx+offset].strip()+" "+keepalpha(word[idx+1:]).strip()
    return word.strip()

def keepalphaS(word,short):
    if short:
        return keepalpha(word)
    else:
        return " ".join(x for x in keepalpha(word).split(" ") if len(x)>=3).upper()
```

```python
def inv(morph):
    ls = morph.split("][")
    cpt=1
    while(cpt>0):
        cpt=0
        for idx,elem in enumerate(ls):
            elem=elem.replace("[","").replace("]","")
            ls[idx]=elem
            if len(elem)<3:
                if idx>0 and elem!='af':
                    ls[idx-1]=ls[idx-1]+ls[idx]
                    del ls[idx]
                    cpt+=1
                else:
                    try:
                        ls[idx+1]=ls[idx]+ls[idx+1]
                        del ls[idx]
                    except:
                        return ls
    sufs=['ing','aar','tor','tje','tie','ator','atie','atje']
    for suf in sufs:
        try:
            i=ls.index(""+suf+"]")
            ls[i-1]=ls[i-1]+""+suf+""
            del ls[i]
        except:
            pass

        try:
            i=ls.index(""+suf+"s]")
            ls[i-1]=ls[i-1]+""+suf+"s"
            del ls[i]
        except:
            pass

        try:
            i=ls.index(""+suf+"s")
            ls[i-1]=ls[i-1]+""+suf+"s"
            del ls[i]
        except:
            pass

        try:
            i=ls.index(""+suf+"")
            ls[i-1]=ls[i-1]+""+suf+""
            del ls[i]
        except:
            pass
    
    return list(ls)
```
